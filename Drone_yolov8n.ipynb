```python
import torch
from pathlib import Path
from PIL import Image
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Define paths
model_path = Path('/content/best.pt')  # Path to your best weights
img_dir = Path('/content/sample_data/daat')   # Path to your image directory

# Load model
model = YOLO(model_path)

# Process 10 images
for i, img_path in enumerate(list(img_dir.glob('*.jpg'))[:10]):  # Process up to 10 images
  print(f"Processing image {i+1}: {img_path}")

  try:
    # Load image
    img = Image.open(img_path)

    # Make prediction
    results = model(img)  # Perform human detection

    # Print results (directly access boxes and confidence scores)
    print("Detected Humans:")
    for detection in results:
      print(f"Bounding Box: {detection[0:4]} Confidence Score: {detection[4]}")

    # Visualize detections (optional)
    fig, ax = plt.subplots()
    ax.imshow(img)

    # Plot bounding boxes (adjust based on your box format)
    for detection in results:
      boxes = detection.boxes.xyxy.cpu()  # Move boxes to CPU
      for box in boxes:
        x_min, y_min, x_max, y_max = box  # Extract coordinates from each box
        rect = plt.Rectangle((x_min, y_min), x_max - x_min, y_max - y_min, color='red', fill=False)
        ax.add_patch(rect)

    plt.show()

  except Exception as e:
    print(f"Error processing image {img_path}: {e}")  # Handle potential errors gracefully
```

```python
import torch
from pathlib import Path
from PIL import Image
from ultralytics import YOLO
import matplotlib.pyplot as plt
import numpy as np
import os  # Added for creating output folder

# Define paths
model_path = Path('/content/best.pt')  # Path to your best weights
img_dir = Path('/content/sample_data/daat')   # Path to your image directory
output_dir = Path('/content/sample_data/output')  # Output directory to save images

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)  # This ensures the folder is created if it doesn't exist

# Load model
model = YOLO(model_path)

# Process images
images = []
results_list = []
for img_path in list(img_dir.glob('*.jpg'))[:25]:  # Process up to 10 images
    img = Image.open(img_path)
    images.append(img)
    results = model(img)  # Perform human detection
    results_list.append(results)

# Function to save image with bounding boxes
def save_image_with_boxes(img, results, output_filename):
    fig, ax = plt.subplots(figsize=(8, 6))  # Adjust figure size as needed
    ax.imshow(img)
    ax.set_title(f"Image with Detections")
    ax.axis('off')

    # Process detections for this image
    for detection in results:
        boxes = detection.boxes.xyxy.cpu()
        for box in boxes:
            x_min, y_min, x_max, y_max = box
            rect = plt.Rectangle((x_min, y_min), x_max - x_min, y_max - y_min, color='red', fill=False)
            ax.add_patch(rect)

    # Save the figure as an image
    fig.savefig(output_dir / output_filename, bbox_inches='tight', pad_inches=0)
    plt.close(fig)  # Close the figure to avoid memory leaks

# Process and save images
for i, (img, results) in enumerate(zip(images, results_list)):
    output_filename = f"output_image_{i+1}.jpg"  # Generate unique filenames
    save_image_with_boxes(img, results, output_filename)

print("Output images saved to:", output_dir)
```
